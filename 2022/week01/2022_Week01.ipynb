```python
import numpy as np
import pandas as pd
import datetime
```

```python
df = pd.read_csv('input/PD 2022 Wk 1 Input - Input.csv', parse_dates=['Date of Birth'])
```

```python
df.head()
```

```python
# Form the pupil's name correctly for the records in the format Last Name, First Name (help)
df["Pupil's name"] = df['pupil last name'] + ', ' +df['pupil first name']
# Form the parental contact's name in the same format as the pupil's 
df['Parental Concat Contact']  = np.where(df['Parental Contact'] == 1, df['Parental Contact Name_1'], df['Parental Contact Name_2'])

# Create the email address to contact the parent using the format Parent First Name.Parent Last Name@Employer.com
df['Parental Concat name'] = df['pupil last name'] + ', ' + df['Parental Concat Contact']
df['Parental Concat Email Adredd'] = df['Parental Concat Contact'] + '.' + df['pupil last name'] + '@'\
                                        + df['Preferred Contact Employer'] + '.com'
```

```python
df.head()
```

```python
# Create the academic year the pupils are in (help)
# Each academic year starts on 1st September.
# Year 1 is anyone born after 1st Sept 2014 
# Year 2 is anyone born between 1st Sept 2013 and 31st Aug 2014 etc
df['Academic Year'] = np.where(df['Date of Birth'] > '2014-09-01', 1, (2015 - df['Date of Birth'].dt.year)
                               + np.where(df['Date of Birth'].dt.month < 9, 1, 0))
```

```python
df.head()
```

```python
col = ['Academic Year', "Pupil's name", 'Parental Concat name', 'Parental Concat Email Adredd']
df.to_csv('output/output-2022-week01.csv', columns=col, index=False)
```
