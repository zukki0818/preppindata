```python
import pandas as pd
```

```python
df = pd.read_csv('input/PD 2022 Wk 5 Input.csv')
df = pd.melt(df, id_vars='Student ID', var_name='Subject', value_name='Score')
df.head()
```

```python
score_map ={'A':10, 'B':8, 'C':6, 'D':4, 'E':2, 'F':1}

df['Grade'] = pd.qcut(df['Score'], q=6, labels=['F', 'E', 'D', 'C', 'B','A'])
df['Points'] = df['Grade'].replace(score_map)
df.head()
```

```python
df['Total Points per Stundent'] = df.groupby(by='Student ID')['Points'].transform('sum')
df.head()
```

```python
df['Avg student total points per grade'] = df.groupby(by='Grade')['Total Points per Stundent'].transform('mean').round(2)
df.head()
```

```python
# Take the average total score you get for students who have received at least one A 
# and remove anyone who scored less than this. 
# Remove students who received an A
at_least_a_avg = df[df['Grade']=='A'].groupby('Student ID')['Total Points per Stundent'].min().mean()
df = df.loc[df['Total Points per Stundent'] > at_least_a_avg]
df = df[df['Grade'] != 'A']
```

```python
#How many students scored more than the average and never received an A?
df['Points per Student without As'] = df.groupby('Student ID')['Points'].transform('sum')
df[df['Points per Student without As'] > at_least_a_avg]['Student ID'].nunique()
```
