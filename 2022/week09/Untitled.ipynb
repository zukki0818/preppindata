```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('input/Sample - Superstore.xls', sheet_name='Orders')
```

```python
df.rename(columns=lambda x: x.replace(' ', '_'), inplace=True)
```

```python
df.head()
```

```python
order = df[['Row_ID', 'Order_Date', 'Customer_ID']]
```

```python
order.head()
```

```python
order['Year'] = order['Order_Date'].dt.year
```

```python
order.head()
```

```python
order['Frist_Purchase'] = order.groupby(by='Customer_ID')['Year'].transform('min')
```

```python
order.head()
```

```python
order['Year'].drop_duplicates()
```

```python
for key, group in df.groupby('Customer_ID'):
    group['Frist_Purchase'].iloc[0]
```
