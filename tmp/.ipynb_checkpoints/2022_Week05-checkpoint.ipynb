```python
import pandas as pd
import numpy as np

df = pd.read_csv('data/PD 2022 WK 3 Grades.csv')
df = pd.melt(df, id_vars='Student ID', var_name='Subject', value_name='Score')
df.head()
```

```python
score_map ={'A':10, 'B':8, 'C':6, 'D':4, 'E':2, 'F':1}

df['Grade'] = pd.qcut(df['Score'], q=6, labels=['F', 'E', 'D', 'C', 'B','A'])
df['Points'] = df['Grade'].replace(score_map)
df
```
