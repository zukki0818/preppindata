```python
import numpy as np
import pandas as pd
```

```python
df_complaints = pd.read_excel('PD 2021 Wk 52 Input.xlsx', sheet_name=0)
keywords = pd.read_excel('PD 2021 Wk 52 Input.xlsx', sheet_name=1)
keywords
```

```python
df_complaints['complaints per customer'] = df_complaints.groupby(['Name']).transform('count') # how to use transform
df_complaints
```

```python
keywords = keywords.append({'Keyword':'other', 'Department':'Unknown'}, ignore_index=True)
keywords
```

```python
keywords_dict = dict(zip(keywords['Keyword'], keywords['Department'])) # how to use zip
keywords_dict
```

```python
df_complaints['complaints causes'] = [[i for i,v in keywords_dict.items() if i.lower() in k] for k in df_complaints['Complaint']]
```

```python
df_complaints = df_complaints.explode('complaints causes')
df_complaints.fillna('other', inplace=True)
```

```python
df_complaints = df_complaints.merge(keywords, how='left', left_on='complaints causes', right_on='Keyword')
df_complaints.head()
```

```python
df_complaints['complaints causes'] = df_complaints.groupby(['Name', 'Complaint', 'complaints per customer', 'Department'])['complaints causes'].transform(lambda x:', '.join(x))
```

```python
df_complaints.head()
```

```python
df_complaints.drop(columns='Keyword', inplace=True)
```

```python
df_complaints.drop_duplicates(inplace=True)
```
