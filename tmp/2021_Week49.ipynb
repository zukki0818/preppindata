```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('data/PD 2021 Wk 49 Input - Input.csv', parse_dates=['Date'])

df['Reporting Year'] = df['Date'].dt.year
df['Reporting Month'] = df['Date'].dt.month

df_out= df.groupby(['Name', 'Reporting Year']).agg(min_date=('Date', 'min'),
                                                 max_date=('Date', 'max'),
                                                 months_worked=('Date', 'count'),
                                                 annual_salary=('Annual Salary', 'mean'),
                                                 total_salary=('Sales', 'sum')).reset_index()

df_out['Salary Paid'] = (df_out['annual_salary'] / 12 * df_out['months_worked']).round(2)
df_out['Yearly Bonus'] = (df_out['total_salary'] * 0.05).round(2)
df_out['Total Paid'] = df_out['Salary Paid'] + df_out['Yearly Bonus']

df_out['Tenure by End of Reporting Year'] = df_out.groupby('Name')['months_worked'].cumsum()
df_out

df_out['Employment Range'] = df_out.groupby(['Name'])['min_date'].transform('min').dt.strftime('%b %Y')\
                            + ' to ' \
                            + df_out.groupby(['Name'])['min_date'].transform('max').dt.strftime('%b %Y')\

df_out.rename(columns={'annual_salary': 'Anual Salary'})
```

```python
col =['Name', 'Employment Range', 'Reporting Year', 'Tenure by End of Reporting Year', 'Salary Paid', 'Yearly Bonus', 'Total Paid']
df_out.to_csv('output-2021-week49.csv', columns=col, index=False)
```
